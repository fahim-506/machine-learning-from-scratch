```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

## 1. Categorical Data
#### a. Countplot

```python
sns.countplot(df['Embarked'])
```

```python
df['Embarked'].value_counts().plot(kind='bar')
```

#### b. PieChart

```python
df['Sex'].value_counts().plot.pie(autopct='%.2f')
```

## 2. Numerical Data
#### a. Histogram

```python
plt.hist(df['Age'],bins=5)
```

#### b. Distplot

```python
sns.distplot(df['Age'])
```

#### c. Boxplot

![Box and Whisker Plot Diagram](https://cdn1.byjus.com/wp-content/uploads/2020/10/Box-Plot-and-Whisker-Plot-1.png)
